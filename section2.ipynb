# Transformada de Fourier

Si $f(x)$ esta definida en el intervalo $[-L,L]$, nosotros somos capaces de representar $f(x)$ como una serie fe Fourier en el intervalo. Ahora si $f$ esta definida en $\mathbb{R}$ y no es periódica, la idea es de extender el concepto de las series de Fourier, para ello usamos la integral.

En esta introducción usaremos conceptos intuitivos, para ello supondremos que $f(x)$  es absolutamente integrable en la recta real, es decir:
$$\int_{-\infty}^\infty|f(x)|dx<\infty,$$
además que $f(x)$ es continua por partes en el intervalo $[-L,L]$, de esta forma 

$$\frac{1}{2L}\int_{-L}^Lf(\xi)d\xi+\sum_{n=1}^\infty\Big[\Big(\frac{1}{L}\int_{-L}^Lf(\xi)\cos\Big(\frac{n\pi\xi}{L}\Big)d\xi\Big)\cos\Big(\frac{n\pi x}{L}\Big)+\Big(\frac{1}{L}\int_{-L}^Lf(\xi)\sin\Big(\frac{n\pi\xi}{L}\Big)d\xi\Big)\sin\Big(\frac{n\pi x}{L}\Big)\Big].$$

Si tomamos $L\to\infty$  para tener la representación de $f$ en toda la recta real y reescribimos los términos 
$$\omega_n=\frac{n\pi}{L},$$
de esta forma la expresión anterior la podemos reescribir como
$$\frac{1}{2\pi}\Big(\int_{-L}^Lf(\xi)d\xi\Big)\Delta \omega+\frac{1}{\pi}\Big[\Big(\int_{-L}^Lf(\xi)\cos(\omega_n\xi)d\xi\Big)\cos(\omega_n x)+\Big(\int_{-L}^Lf(\xi)\sin(\omega_n\xi)d\xi\Big)\sin(\omega_n x)\Big]\Delta \omega.$$


Tomando $L\to \infty$, así $[-L,L]$ expandiendo sobre la línea real. Así tenemos que $\Delta \omega\to0$. al examinar la primera ecuación;
$$\frac{1}{2\pi}\Big(\int_{-L}^Lf(\xi)d\xi\Big)\Delta \omega\to0,$$ puesto que la integral converge y es acotada. 
El otro término de la ecuación se puede asemejar como una suma de Riemann. Como 
$L\to \infty$ y $\Delta\omega \to \infty$ esta expresión se aproxima 

$$\frac{1}{2\pi}\Big[\Big(\int_{\infty}^{\infty}f(\xi)\cos(\omega\xi)\Big)\cos(\omega x)d \xi+\Big(\int_{\infty}^{\infty}f(\xi)\sin(\omega\xi)\Big)\sin(\omega x)d \xi\Big]d\omega$$

Asi  en la interal de Fourier de $f(x)$ en la linea real esta definida de la forma 
$$\int_0^\infty\Big(A_\omega\cos(\omega x)+B_\omega\sin(\omega x)\Big)d\omega,$$
donde 
$$A_\omega=\frac{1}{\pi}\int_{-\infty}^\infty f(\xi)\cos(\omega\xi)d\xi,$$
y
$$B_\omega=\frac{1}{\pi}\int_{-\infty}^\infty f(\xi)\sin(\omega\xi)d\xi .$$

La integral con respecto a $\omega$ reempla la sumatoria con respecto a $n$


### Teorema

Supongamos que $f(x)$ está definida para todo realxy que $\int_{-\infty}^\infty|f(x)|dx$ converge. Suponga que es suave por partes en cada intervalo $[−L,L]$ para $L>0$. Entonces, en cualquier $x$, la integral de Fourier  converge a
$$\frac{(f(x+)+f(x−))}{2}$$. En particular, si $f$ es continua en $x$, entonces la integral de Fourier converge para todo $x$ a $f(x)$.

## transformada de Fourier
La transformada de Fourier de una función $f$ se define como

$$F(\omega) = \int_{-\infty}^\infty f(x) e^{-2\pi i x \omega} \mathrm{d} x.$$

## Cálculo de la transformada de Fourier

Para calcular las integrales podemos usar la libreria [Sympy](https://docs.sympy.org/latest/modules/integrals/integrals.html)


```python
from sympy import *
init_printing()
```

creamos algunas variables simolicas 

```python
x, k, xi, t, omega = symbols('x, k, xi, t, omega', real=True)
a, k0, alpha = symbols('a, k_0, alpha', positive=True)
```

### transformadad e la Gausiana

$f(t) = e^{-at^2}$

```python
f1 = exp(-a*t**2)  #d efinimos la función
fourier_transform(f1,t,omega/(2*pi))
```

## Sinusoidal amorgtiuada con función gaussiana
$f(t)= e^{-at^2}cos(k_0t)$

```python
f2 = exp(-a*t**2)*cos(k0*t)
Tf2=fourier_transform(f2,t,k/(2*pi))
Tf2
```

La expresión anterior se puede simplificar

```python
simplify(Tf2.as_real_imag())[0]
```

### Función por partes

```python
f3 = Heaviside(x+a)*Heaviside(a-x)
Tf3 = fourier_transform(f3,x,k/(2*pi))
simplify(Tf3.as_real_imag())[0]
```

### Delta dirac

```python
f4 = DiracDelta(x-xi)

fourier_transform(f4,x,k/(2*pi))
```
