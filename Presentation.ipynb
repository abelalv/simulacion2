# Modelación y Simulacion II


Las ecuaciones diferenciales parciales son de vital importancia para modelar muchos fenómenos físicos, biológicos, epidemiológicos, entre otros. Por tal motivo se hace indispensable en un programa de matemática aplicada ver una introducción de los diferentes fenómenos que modelan este tipo de ecuaciones y algunas herramientas matemáticas para la simulación de estos. 


## Objectivos 

+ Al finalizar este curso el estudiante debe adquirir la habilidad de modelar fenómenos de variación temporal a través de modelos determinísticos usando ecuaciones diferenciales parciales.  

+ Aproximar adecuadamente ecuaciones diferenciales parciales, por medio de diferencias finitas, para fenómenos difusivos, advectivos y advectivos-difusivos, lineales. 

## Contenido 


 
* Introducción a la teoría de Fourier. Series Fourier, Transformada de Fourier, Transformada rápida de Fourier Transformada de Fourier aplicadas a la EDP. 


* Introducción, Ecuaciones de primer orden, Ecuaciones de segundo orden, clasificación. 


* Modelización en ecuaciones diferenciales parciales.  Ecuaciones de mecánica del continuo. Coordenadas Eulerianas-Lagrangianas. Teorema del transporte, leyes de conservación (ecuaciones diferenciales parciales). Ecuación de convección. Ecuación de conducción del calor, proceso de difusión. Ecuación de convección-difusión. 

* Métodos numéricos para ecuaciones parabólicas. Método de Euler Explicito. Método semidescrizado, método de las líneas, condición CFL. Métodos implícitos. Análisis de Von-Neuwman. Problemas de valor inicial. Condiciones de frontera mixtas. Métodos para dos dimensiones, método splitting, Método ADI. Programación. Métodos mixtos (explicito/implícito) ecuación de advección difusión. Aplicaciones reales. 

 

* Métodos numéricos para ecuaciones diferenciales hiperbólicas, Método de las líneas, Métodos de segundo orden en el espacio. Análisis de Von-Neuwman. Condición CFL. Sistemas hiperbólicos. Sistemas hiperbólicos lineales. Métodos para sistemas hiperbólicos. Aplicaciones reales. 

 
* Métodos para ecuaciones elípticas. Discretización y condiciones acotas.  Principio del máximo, Estimaciones del error global. Operador splitting. Aplicaciones reales.
